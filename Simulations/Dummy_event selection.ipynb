```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, BoundaryNorm
from matplotlib.patches import Rectangle
```

```python
matrix1=np.zeros((30,30))


matrix1[12:17,25]=37

matrix1[11:14,1]=31
matrix1[12:15,2]=33
matrix1[13:17,3]=32
matrix1[14:18,4]=33
matrix1[15:20,5]=34
matrix1[16:21,6]=35
matrix1[17:22,7]=35
matrix1[18:23,8]=34


matrix1+=np.random.randint(25,35,(30,30))




matrix2=np.zeros((30,30))


matrix2[10:16,25]=37

matrix2[5:16,3]=32
matrix2[10:18,4]=33
matrix2[8:19,5]=34
matrix2[12:21,6]=35
matrix2[10:22,7]=35
matrix2[15:23,8]=34
matrix2[14:24,9]=34
matrix2[18:25,10]=33
matrix2[21:26,11]=31

matrix2+=np.random.randint(25,35,(30,30))





matrix3=np.zeros((30,30))


matrix3[10:17,25]=37

matrix3[7:17,3]=32
matrix3[9:18,4]=33
matrix3[13:20,5]=34
matrix3[15:21,6]=35
matrix3[17:22,7]=35
matrix3[18:23,8]=34
matrix3[19:24,9]=34
matrix3[21:25,10]=33
matrix3[22:26,11]=31

matrix3+=np.random.randint(25,35,(30,30))
```

```python

fig, axes = plt.subplots(nrows=1,ncols=3, figsize=(12, 2), sharex=True)

matrices = [matrix1, matrix2, matrix3]
titles = ['RTI Matrix Ch.0', 'RTI Matrix Ch.1', 'RTI Matrix Ch.2']  # Opcional, para identificar cada panel

for ax, matrix, title in zip(axes, matrices, titles):
    cax = ax.imshow(matrix, cmap='jet', vmin=0, vmax=100, aspect='auto')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylabel('Range', fontsize=17)
    ax.set_title(title, fontsize=12)  # Descomenta si quieres un título en cada subplot
    ax.set_xlabel('Time', fontsize=17)


# Colorbar común a todos los subplots
cbar = fig.colorbar(cax, ax=axes.ravel().tolist(), shrink=0.95)
cbar.set_label('Power', fontsize=17)
cbar.set_ticks([])

#lt.tight_layout()
plt.show()
```

```python
matrix1=matrix1>35
matrix2=matrix2>35
matrix3=matrix3>35
```

```python
cmap = ListedColormap(['lightgray', 'blue'])
norm = BoundaryNorm([0, 0.5, 1], cmap.N)  # Límite en 0.5 entre False (0) y True (1)

fig, axes = plt.subplots(nrows=1,ncols=3, figsize=(12, 2), sharex=True)

matrices = [matrix1, matrix2, matrix3]
titles = ['Boolean Matrix Ch.0', 'Boolean Matrix Ch.1', 'Boolean Matrix Ch.2']  # Opcional

for ax, matrix, title in zip(axes, matrices, titles):
    cax = ax.imshow(matrix, cmap=cmap, norm=norm, aspect='auto')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylabel('Range', fontsize=17)
    ax.set_title(title, fontsize=11)
    ax.set_xlabel('Time', fontsize=17)

# Colorbar con dos etiquetas: "False" y "True"
cbar = fig.colorbar(cax, ax=axes.ravel().tolist(), ticks=[0.25, 0.75], shrink=0.95)
cbar.set_ticklabels(['False', 'True'])
cbar.set_label('Significance', fontsize=17)


plt.show()
```

```python
coincident_matrix=matrix1*matrix2*matrix3
```

```python


# Definir el mapa de colores binario
cmap = ListedColormap(['lightgray', 'blue'])
norm = BoundaryNorm([0, 0.5, 1], cmap.N)

fig, ax = plt.subplots(figsize=(4, 2.5))  # Ajusta altura para un solo gráfico

cax = ax.imshow(coincident_matrix, cmap=cmap, norm=norm, aspect='auto')
ax.set_xticks([])
ax.set_yticks([])
ax.set_ylabel('Range', fontsize=17)
ax.set_xlabel('Time', fontsize=17)
ax.set_title('Coincidence Boolean Matrix', fontsize=11)

# Colorbar con dos etiquetas
cbar = fig.colorbar(cax, ax=ax, ticks=[0.25, 0.75], shrink=0.95)
cbar.set_ticklabels(['False', 'True'])
cbar.set_label('Coincidence', fontsize=17)

plt.tight_layout()
plt.show()
```

```python
# Definir el mapa de colores binario
cmap = ListedColormap(['lightgray', 'blue'])
norm = BoundaryNorm([0, 0.5, 1], cmap.N)

fig, ax = plt.subplots(figsize=(4, 2.5))  # Ajusta altura para un solo gráfico

# Mostrar la matriz
cax = ax.imshow(coincident_matrix, cmap=cmap, norm=norm, aspect='auto')

# Dibujar el rectángulo para coincident_matrix[10:16, 25]
rect = Rectangle(
    (25 - 1, 10 - 1),  # esquina inferior izquierda (x, y)
    width=2,               # una sola columna
    height=8,              # de la fila 10 a la 15 (inclusive) → 6 filas
    edgecolor='red',
    facecolor='none',
    linewidth=1
)
ax.add_patch(rect)

# Ajustar etiquetas y título
ax.set_xticks([])
ax.set_yticks([])
ax.set_ylabel('Range', fontsize=17)
ax.set_xlabel('Time', fontsize=17)
ax.set_title('Shape Filter', fontsize=11)

# Colorbar con dos etiquetas
cbar = fig.colorbar(cax, ax=ax, ticks=[0.25, 0.75], shrink=0.95)
cbar.set_ticklabels(['False', 'True'])
cbar.set_label('Coincidence', fontsize=17)

plt.tight_layout()
plt.show()
```
